```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/planets.csv"
df = pd.read_csv(url)

print(df.head())
```

```python
df_numerico = df[["year", "orbital_period", "mass"]]
print(df_numerico.head())
```

```python
import matplotlib.pyplot as plt
```

```python
df["year"].hist(bins=20)
plt.title("Descubrimiento de planetas por a√±o")
plt.xlabel("A√±o")
plt.ylabel("Cantidad")
plt.show()
```

```python
plt.scatter(df["orbital_period"], df["mass"])
plt.xlabel("Periodo orbital")
plt.ylabel("Masa")
plt.title("Periodo orbital vs Masa de planetas")
plt.show()
```

```python
df_clean = df[["orbital_period", "mass"]].dropna()
print(df_clean.head())
```

```python
from sklearn.cluster import KMeans

X = df_clean

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

df_clean["grupo"] = kmeans.labels_
print(df_clean.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    df_clean["orbital_period"],
    df_clean["mass"],
    c=df_clean["grupo"]
)
plt.xlabel("Periodo orbital")
plt.ylabel("Masa")
plt.title("Clustering de planetas (IA)")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[["orbital_period", "mass"]])

kmeans = KMeans(n_clusters=3, random_state=42)
df_clean["grupo_normalizado"] = kmeans.fit_predict(X_scaled)
```

```python
plt.scatter(
    df_clean["orbital_period"],
    df_clean["mass"],
    c=df_clean["grupo"],
    cmap="viridis"
)
plt.show()
```
